## TASK:1

Task: Restaurant Reviews

Analyze the text reviews to identify the most
common positive and negative keywords.
Calculate the average length of reviews and
explore if there is a relationship between
review length and rating.

```python
import pandas as pd
import numpy as np
```

```python
file_path = r"C:\Users\riyap\Downloads\Dataset .csv"

df = pd.read_csv(file_path)
```

```python
df = df.dropna(subset = ["Rating text", "Aggregate rating"])

df["Rating text"] = df["Rating text"].str.lower()
```

```python
all_words = df["Rating text"].str.split().explode()

word_counts = all_words.value_counts()
```

```python
print("most common review keywords:--")
print(word_counts)
print()
```

```python
df["Review Length"] = df["Rating text"].apply(len)

avg_length = df["Review Length"].mean()
print("Average review length:--", round(avg_length, 2))
print()
```

```python
rating_vs_length = df.groupby("Rating text")["Aggregate rating"].mean()

print("Average rating based on review text:--")
print(rating_vs_length)
```

```python
## CONCLUSION....

# Restaurants labeled "excellent" have the highest average ratings.
# "Very good" restaurants also show consistently high ratings.
# Rating decrease steadily from good to average to poor.
# The "not rated" category shows a zero rating and does not reflect actual performance.
# Text-based rating align well with numerical aggregate ratings.
```

## TASK:2

Task: Votes Analysis

Identify the restaurants with the highest and
lowest number of votes.
Analyze if there is a correlation between the
number of votes and the rating of a
restaurant.

```python
df = df.dropna(subset=["Votes", "Aggregate rating"])
```

```python
highest_votes = df["Votes"].max()
lowest_votes = df["Votes"].min()
```

```python
high_vote_restaurant = df[df["Votes"] == highest_votes]
low_vote_restaurant = df[df["Votes"] == lowest_votes]
```

```python
print("Restaurant with highest votes:---")
print(high_vote_restaurant[["Restaurant Name", "Votes"]])
print()
```

```python
print("Restaurant with lowest votes:---")
print(low_vote_restaurant[["Restaurant Name", "Votes"]])
print()
```

```python
correlation = np.corrcoef(df["Votes"], df["Aggregate rating"])[0][1]

print("Correlation between votes and rating:---", round(correlation, 2))
```

```python
## CONCLUSION

# Restaurants with more votes generally tend to have hidher rating.
# The relationship between votes and rating is positive but not strong.
# High customer engagement does not always guarantee a high rating.
# Some low-voted restaurants still maintain good ratings.
# Popularity and quality are related, but other factors also influence ratings.
```

## TASK:3

Task: Price Range vs. Online Delivery and
Table Booking

Analyze if there is a relationship between the
price range and the availability of online
delivery and table booking.
Determine if higher-priced restaurants are
more likely to offer these services.

```python
df = df[["Price range", "Has Online delivery", "Has Table booking"]]
```

```python
online_delivery = df.groupby("Price range")["Has Online delivery"].value_counts(normalize = True) * 100

print("Online delivery percentage by price range:---")
print(online_delivery)
print()
```

```python
table_booking = df.groupby("Price range")["Has Table booking"].value_counts(normalize=True) * 100

print("Table booking percentage by price range:---")
print(table_booking)
```

```python
## CONCLUSION...

# Higher price range restaurants tend to offer online delivery more frequently.
# Table booking facilities are more common in higher_priced restaurants.
# Lower-priced restaurants mostly do not provide table booking services.
# Online delivery availability increases gradully with price range.
# Price range appears to influence the type of services offered by restaurants.
```
